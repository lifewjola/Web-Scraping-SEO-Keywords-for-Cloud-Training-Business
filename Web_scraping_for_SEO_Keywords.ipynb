```python
from bs4 import BeautifulSoup

file_path = '/content/SEO Monitoring.html'
```

### My HTML Knowledge came in handy in inspecting the website

After reading the file, I had to inspect the html file in the browser to know the tags and the classes/ id's of the tags that contained our juice (I mean the keywords ><)

```python
with open(file_path, 'r', encoding='utf-8') as file:
    content = file.read()

soup = BeautifulSoup(content, "html.parser")
keywords = soup.find_all("div", class_="swTable-keywordCell u-relative u-full-width")
```

### Soooo...
My code was looking too simple so I decided to save the keywords in a CSV file (you know, to make things more interesting around here)

```python
import csv

cloudkeywords = []
for keyword in keywords:
    span_tag = keyword.find("span")
    if span_tag:
        keyword = span_tag.find("a", class_="swTable-content").text
        cloudkeywords.append(keyword)
output_file = 'cloudkeywords.csv'
```

### Ooooh, here comes my favorite part...
time to write some keywords to a CSV

```python
with open(output_file, 'w', newline='', encoding='utf-8') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(['Cloud Keywords \nThe keywords are in descending order of relevance'])
    for keyword in cloudkeywords:
        writer.writerow([keyword])

print(f"Successfully saved {len(cloudkeywords)} keywords to {output_file}.")
```
